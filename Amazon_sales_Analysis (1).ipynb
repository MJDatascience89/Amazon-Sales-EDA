```python
def remove_outliers_iqr(df, column_name):  #Removes outliers from a specified column of a Pandas DataFrame using the Interquartile Range (IQR) method.

    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_no_outliers = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    #price_counts = df_no_outliers[column_name].value_counts(bins=5) # Adjust number of bins as needed

    plt.figure(figsize=(8, 6))
    plt.hist(df_no_outliers[column_name], bins=10)
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
#for i in range(len(price_counts)):
  #    plt.text(price_counts.index[i].left, price_counts.values[i], str(price_counts.values[i]), ha='center', va='bottom')

    return df_no_outliers


numerical_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating_count', 'rating']

for col in numerical_cols:
    df = remove_outliers_iqr(df,col)

```

*   'Discounted Price' distribution is right-skewed indicates most product have lower discounted price.Where Discount percentage distribution lies between 40 -80.
*   'rating count' distribustion shows some product have higher number of rating.

```python
import re

def preprocess_text(text):
    text = text.lower()  # Lowercase the text
    text = re.sub(r'[^\w\s]', '', text) # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespace
    text = text.replace('usb cable', 'usbcable')
    text = text.replace('type c', 'typec')
    text = text.replace('fast charging', 'fastcharging')
    text = text.replace('black cable', 'blackcable')
    text = text.replace('data cable', 'datacable')
    text = text.replace('white cable', 'whitecable')

    return text

df['product_name'] = df['product_name'].apply(preprocess_text)
df['review_content'] = df['review_content'].apply(preprocess_text)
```

```python
import nltk #natural language toolkit library
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
df['sentiment_score'] = df['review_content'].apply(lambda review: sia.polarity_scores(review)['compound'])
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['sentiment_score'], bins=20)  # Adjust bins as needed
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

Sentiment score more towards 1.0 means it goes  most with positive review.

```python
positive_reviews = df[df['sentiment_score'] > 0.05]['product_name'].count()
negative_reviews = df[df['sentiment_score'] < -0.00]['product_name'].count()
neutral_reviews = df[(df['sentiment_score'] >= -0.00) & (df['sentiment_score'] <= 0.05)]['product_name'].count()

print(f"Positive reviews: {positive_reviews}")
print(f"Negative reviews: {negative_reviews}")
print(f"Neutral reviews: {neutral_reviews}")
```

**Positive review are much higher than Negative review**
